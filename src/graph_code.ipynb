```python
import matplotlib.pyplot as plt
import math
import json
import numpy as np

with open('/Users/rahulbalakrishnan/Desktop/throttlebot/src/performance_results_provision2', 'r') as file:
    performance_data = json.loads(file.read())
    
with open('/Users/rahulbalakrishnan/Desktop/throttlebot/src/cost_results_provision2', 'r') as file:
    cost_data = json.loads(file.read())

utilization_performance = []
avg_latency_data = []
for utilization in performance_data: 
    utilization_performance.append(utilization['utilization'])
    data = utilization['data']
    
    sum_value_to_measure = 0 
    for pod in data: 
        value_to_measure = pod['latency_99']
        sum_value_to_measure += float(value_to_measure[0])
    
    avg_latency = sum_value_to_measure / len(data)
    avg_latency_data.append(avg_latency)

utilization_cost = []
cost = []
for utilization in cost_data: 
    utilization_cost.append(utilization['utilization'])
    
    data = utilization['data']
    cost.append(data)


print(avg_latency_data)
print(cost)
print(avg_latency_data2)
print(cost2)
```

```python
fig, ax1 = plt.subplots()

color = 'red'
color2 = 'purple'

ax1.set_xlabel('utilization')
ax1.set_ylabel('cpu-sec', color=color)
ax1.plot(utilization_cost, cost, color=color, label = "intial cost")
ax1.plot(utilization_cost, cost2, color = color2, label = "cost after increased workload")
ax1.tick_params(axis='y', labelcolor=color)
plt.legend(loc = "upper right")

# ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

fig, ax2 = plt.subplots()
color = 'green'
color2 = 'cyan'

ax2.set_ylabel('latency_99', color=color)  # we already handled the x-label with ax1
ax2.plot(utilization_cost, avg_latency_data, color=color, label ="initial latency")
ax2.plot(utilization_cost, avg_latency_data2, color = color2, label = "latency after increased workload")
ax2.tick_params(axis='y', labelcolor=color)
plt.legend(loc="upper right")



fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
    
    
```
